```python
import pandas as pd
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df.head()
```

```python
# inputs.describe()
```

```python
df.drop(["Name","Ticket","PassengerId","SibSp","Parch","Cabin","Embarked"],axis="columns",inplace=True)
```

```python
df.head()
```

```python
inputs = df.drop('Survived',axis='columns')
target=df["Survived"]
```

```python
inputs.Sex=inputs.Sex.map({"male":1,"female":0})
```

```python
inputs.head()
```

```python
inputs.isna().sum()
```

```python
inputs.Age=inputs.Age.fillna(inputs.Age.mean())
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(inputs,target,test_size=0.2)
```

```python
#X is used for input data
#Y is used for output data 
#X_train is input training data
#X_test is input testing data
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

```python
## using criteria as entropy
```

```python
entropic_model=tree.DecisionTreeClassifier(criterion='entropy')
```

```python
entropic_model.fit(X_train,Y_train)
```

```python
entropic_model.score(X_test,Y_test)
```

```python
prediction=entropic_model.predict([[1,0,38.0,71.2833]])
```
